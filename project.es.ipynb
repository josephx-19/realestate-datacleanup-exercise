#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
# TODO
name1= ds.loc[ds["level5"] == "Valdemorillo"]
price1 = name1["price"]

name2= ds.loc[ds["level5"] == "Galapagar"]
price2 = name2["price"]

print("precios promedios de Valdemorillo:",np.mean(price1))
print("precios promedios de Galapagar:",np.mean(price2))

if np.mean(price1)/np.mean(price2) == 1 :
    print("al dividir los precios de Valdemorillo y Galapagar el resultado es 1 con lo cual es exactamente igual" )
elif np.mean(price1)/np.mean(price2) > 1:
    print("El promedio de precios de Valdemorillo es mas alto que Galapagar")
else:
    print("El promedio de precios de Galapagar es mas alto que Valdemorillo")
```
